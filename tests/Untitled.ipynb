```python
def read_keypoints():
    keypoints = []
    for i in range(0, 100):
        filename = '/home/paul/git/Splines/data/kinect_paper/frame%d.txt' % i
        keypoints.append(np.loadtxt(filename))
    return keypoints

def read_images():
    images = []
    for i in range(1, 101):
        filename = '/home/paul/git/Splines/data/kinect_paper/images/%03d.bmp' % i
        images.append(cv2.imread(filename))
    return images

def draw_keypoints(img, kpts):
    img = img.copy()
    for i in range(kpts.shape[1]):
        cv2.circle(img,(int(kpts[0,i]), int(kpts[1,i])), 3, (0,255,0), -1)
    return img
```
